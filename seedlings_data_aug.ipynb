```python
def get_train_data_frame(path):
    print(path)
    classes = check_output(["ls", path]).decode("utf8").strip().split("\n")
    print(classes)
    dir_list = []
    
    for c in classes:
        files = check_output(["ls", "%s%s" % (path, c)]).decode("utf8").strip().split("\n")
        dir_list.append(files)
        
    images = []
    paths = []
    im_class = []
    im_height = []
    im_width = []
    for c, files in zip(classes, dir_list):
        for img in files:
            img_path = "%s%s/%s" % (path, c, img)
            paths.append(img_path)
            im = Image.open(img_path)
            images.append(img)
            im_class.append(c)
            im_height.append(im.height)
            im_width.append(im.width)

    df_all = pd.DataFrame({"class": im_class, "height": im_height, "width": im_width}, index=images)
    
    df_by_class = pd.DataFrame({"file": images, 
                                "class": im_class, 
                                "height": im_height, 
                                "width": im_width, 
                                "path": paths}, 
                               index=paths)

    return df_by_class
```
